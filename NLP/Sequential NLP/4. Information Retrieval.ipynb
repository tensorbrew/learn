#### Removing Personal information

```
import spacy
```

```
# Load the large English NLP model
nlp = spacy.load('en_core_web_lg')
```

```
#Example text
text1 = 'My name is John.'
```

```
#Parse the text using spaCy
doc = nlp(text1)
```

```
#Visualize dependecy parsing 
from spacy import displacy
```

```
displacy.render(doc, style="ent", jupyter=True)
```

```
#Check Named entity Recognition
for token in doc:
    print(token, token.ent_type_)
```

Function to remove name

```
def remove_name(text, replacement_token='[NAME]'):

    #Parse the text
    doc = nlp(text)

    #Updated document
    updated_doc = []

    #Check Entity type
    for token in doc:
        if token.ent_type_ == 'PERSON':
            updated_doc.append(replacement_token)
        else:
            updated_doc.append(str(token))
    
    return ' '.join(updated_doc)
```

```
remove_name(text1)
```

```
remove_name('My name is Rajeev Kumar.')
```

```
displacy.render(nlp('My name is Rajeev Kumar.'), style="ent", jupyter=True)
```

##### How do we correct that?

```
#Lets check how many entities do we have?
doc = nlp('My name is Rajeev Kumar Singh')
i = 1
for ent in doc.ents:
    print(str(i) + '.',ent)
    i += 1
```

```
#How do we check if a token i.e word is beginning of a multiple words entity
for token in doc:
    #Lets check IOB value
    print(token, token.ent_type_, token.ent_iob_)
```

```
def remove_name2(text, replacement_token='[NAME]'):

    #Parse the text
    doc = nlp(text)

    #Updated document
    updated_doc = []

    #Check Entity type
    for token in doc:
        if token.ent_type_ == 'PERSON':
            if token.ent_iob_ == 'B':
                #Replace starting entity word
                updated_doc.append(replacement_token)
            else:
                #ignore 
                pass
        else:
            updated_doc.append(str(token))
    
    return ' '.join(updated_doc)
```

```
remove_name2('My name is Rajeev Kumar Singh.')
```
