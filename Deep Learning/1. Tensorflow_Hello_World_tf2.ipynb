### Adding two tensors

```
a = tf.constant([1.2])
b = tf.constant([3.4])
c = tf.add(a,b)
print(c)
```

TensorFlow will internally convert data to Tensor if needed

```
a = [1.2]
b = [3.4]
c = tf.add(a,b)
print(c)
```

```
a = np.array([1.2])
b = np.array([3.4])
c = tf.add(a,b)
print(c)
```
